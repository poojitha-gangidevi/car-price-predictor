```python
import pandas as pd
import re

# 1. Load the file 
car = pd.read_csv('Cleaned Car.csv')

# 2. FIX THE SPACES FIRST: Protect word boundaries, remove letter spaces
def clean_car_name(text):
    text = str(text)
    # Split where there are multiple spaces (word boundaries)
    words = re.split(r'\s{2,}', text)
    # Remove single spaces inside each word and join back with a clean single space
    cleaned_words = [word.replace(' ', '') for word in words]
    return ' '.join(cleaned_words)

car['name'] = car['name'].apply(clean_car_name)

# 3. NOW slice the first 3 words safely
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')

# 4. Update features and re-fit the encoder
X = car.drop(columns='Price')
ohe.fit(X[['name', 'company', 'fuel_type']])

# 5. See the final correct result!
ohe.categories_
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

# 1. Force the encoder to ignore any unseen car names completely
ohe = OneHotEncoder(handle_unknown='ignore')

# 2. Build the transformer and pipeline from scratch
column_trans = make_column_transformer(
    (ohe, ['name', 'company', 'fuel_type']),
    remainder='passthrough'
)
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)

# 3. Train the fresh pipeline on your training data
pipe.fit(x_train, y_train)

# 4. Create the input data frame for prediction
input_data = pd.DataFrame(
    columns=['name', 'company', 'year', 'kms_driven', 'fuel_type'],
    data=np.array(['Maruti Suzuki Swift', 'Maruti', 2019, 100, 'Petrol']).reshape(1, 5)
)

# 5. Predict the price safely
prediction = pipe.predict(input_data)
print(f"Predicted Price: ₹{prediction[0]:,.2f}")
```
